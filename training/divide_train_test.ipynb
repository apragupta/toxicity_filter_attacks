# Notebook to divide the data into train and test data

```python
import pandas as pd 
from random import shuffle,seed
seed(0)
import os
```

```python
folder_path = r'../preprocessed/try_2'
```

```python
#load the comments and sentence dataframes
comments_df_path = os.path.join(folder_path, 'comments_df')
sentence_df_path = os.path.join(folder_path, 'sentence_df')
```

```python
comments_df = pd.read_pickle(comments_df_path)
sentences_df = pd.read_pickle(sentence_df_path)
```

```python
comment_ids = comments_df['comment_idx'].to_list()
```

```python
shuffle(comment_ids)
```

```python
#now get train and test comment indices by simply splitting the shuffled indices 80-20

train_len = int(0.8*len(comment_ids))
```

```python
train_ids = comment_ids[0:train_len]
```

```python
train_comments = comments_df[comments_df.comment_idx.isin(train_ids)]
test_comments = comments_df[~comments_df.comment_idx.isin(train_ids)]
```

```python
train_comments
```

```python
test_comments
```

```python
train_sentences = sentences_df[sentences_df['Comment #'].isin(train_ids)]
test_sentences = sentences_df[~sentences_df['Comment #'].isin(train_ids)]
```

```python
train_sentences
```

```python
test_sentences
```

```python
train_comments.to_pickle(os.path.join(folder_path, 'train_comments_df'))
test_comments.to_pickle(os.path.join(folder_path, 'test_comments_df'))
train_sentences.to_pickle(os.path.join(folder_path, 'train_sentence_df'))
test_sentences.to_pickle(os.path.join(folder_path, 'test_sentence_df'))
```
